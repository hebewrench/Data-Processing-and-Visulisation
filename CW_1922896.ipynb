```python
# CELL3 – 5 marks
Academic_Reputation_mean=round(QS_UK_rankings["Academic Reputation"].mean(),2)
Academic_Reputation_std=round(QS_UK_rankings["Academic Reputation"].std(ddof=0),2)
minimum=round(QS_UK_rankings["Academic Reputation"].min(),2)
maximum=round(QS_UK_rankings["Academic Reputation"].max(),2)
Academic_Reputation_err=1.96*std(QS_UK_rankings['Academic Reputation'], ddof=0)/sqrt(len(QS_UK_rankings['Academic Reputation']))
print()
print("mean = "+ str(Academic_Reputation_mean))
print("minimum = " + str(minimum))
print("maximum = " + str(maximum))
print("standard deviation = "+ str(Academic_Reputation_std))
print('95% Academic Reputation CI: ',[round((Academic_Reputation_mean-Academic_Reputation_err),2), round((Academic_Reputation_mean+Academic_Reputation_err),2)])
```

```python
# CELL4 – 9 marks
# means for citations by country
germany=data[data['Location'] == "Germany"]
ge_cit_mean=germany['Citations per Faculty'].mean()
us=data[data['Location'] == "United States"]
us_cit_mean=us['Citations per Faculty'].mean()
canada=data[data['Location'] == "Canada"]
ca_cit_mean=canada['Citations per Faculty'].mean()
uk=data[data['Location'] == "United Kingdom"]
uk_cit_mean=uk['Citations per Faculty'].mean()
australia=data[data['Location'] == "Australia"]
aus_cit_mean=australia['Citations per Faculty'].mean()
netherlands=data[data['Location'] == "Netherlands"]
ne_cit_mean=netherlands['Citations per Faculty'].mean()


#Confidence intervals by country
ge_err=1.96*std(germany['Citations per Faculty'], ddof=0)/sqrt(len(germany['Citations per Faculty']))
us_err=1.96*std(us['Citations per Faculty'], ddof=0)/sqrt(len(us['Citations per Faculty']))
ca_err=1.96*std(canada['Citations per Faculty'], ddof=0)/sqrt(len(canada['Citations per Faculty']))
uk_err=1.96*std(uk['Citations per Faculty'], ddof=0)/sqrt(len(uk['Citations per Faculty']))
au_err=1.96*std(australia['Citations per Faculty'], ddof=0)/sqrt(len(australia['Citations per Faculty']))
ne_err=1.96*std(netherlands['Citations per Faculty'], ddof=0)/sqrt(len(netherlands['Citations per Faculty']))

#bar graph
figure(figsize(12,6))
N=6
ind = arange(N)
width = 0.5
bar(ind, [ge_cit_mean, us_cit_mean, ca_cit_mean, uk_cit_mean, aus_cit_mean, ne_cit_mean], width, color='r', 
    yerr=[ge_err, us_err, ca_err, uk_err, au_err, ne_err])
ylabel('Mean Citations per Faculty', fontsize=15)
xtickloc = ind
xticklabels = ['Germany', 'United States', 'Canada', 'United Kingdom', 'Australia', 'Netherlands']
xticks(xtickloc, xticklabels, rotation='horizontal', fontsize=15)
title('Average Citations per Faculty by Country', fontsize=22)
show()
```

```python
# CELL7 – 14 marks
q_score=pd.read_excel('./Q_scores.xlsx')
x=[]
y=[]
for i in range(len(q_score)):
    if 'm' in q_score['Image'][i]:
        IQ=q_score['Quality'][i]
        n=i+1
        img = imread('m'+str(n)+'.png')*255
        API=img.mean(axis=0).mean()
        x.append(API)
        y.append(IQ)

scatter(x,y)
xlabel('Average Pixel Intensity', fontsize=15)
ylabel('Image Quality', fontsize=15)
title('Average Pixel Intensity Compared with Image Quality on the model dataset', fontsize=20)

slope, intercept, rvalue, pvalue, stderr = stats.linregress(x, y)
line = slope * np.array(x) + intercept
plot(x, line,'b-', label='regression line')

Yi=[]
y_minus_Yi=[]
for n in x:
    Yi.append(slope*n+intercept)

dif_squared=[]
for n in range(len(y)):
    y_minus_Yi.append(y[n]-Yi[n])
    dif_squared.append(y_minus_Yi[n]*y_minus_Yi[n])
    
sum_diffs=sum(dif_squared)
MSE=sum_diffs/(len(y))
IQscore55 = (5.5 - intercept)/slope

x_vals=[]
y_vals=[]
ang=degrees(178)
line_vals_x=[]
line_vals_y=[]

for n in range(len(y)):
    line_vals_x.append((y[n] - intercept)/slope)
    line_vals_y.append(y[n])
    
for n in range(len(line_vals_x)):
    xnew = IQscore55 + (line_vals_x[n]-IQscore55)*cos(ang) + ((line_vals_y[n]-5.5)*sin(ang))
    ynew = 5.5 + (line_vals_x[n]-IQscore55)*sin(ang) - (line_vals_y[n] - 5.5)*cos(ang)
    x_vals.append(xnew)
    y_vals.append(ynew)


slope2, intercept2, rvalue2, pvalue2, stderr2 = stats.linregress(x_vals, y_vals)
line2 = slope2 * np.array(x_vals) + intercept2
plot(x_vals, line2,'--', label='rotated by 2 degrees')
legend(loc='best', fontsize=15)
show()

Yr=[]
y_minus_Yr=[]
dif_squared_r=[]
for i in x:
    Yr.append(slope2*i+intercept2)

for n in range(len(y)):
    y_minus_Yr.append(y[n]-Yr[n])
    dif_squared_r.append(y_minus_Yr[n]*y_minus_Yr[n])
    
sum_diffs_r=sum(dif_squared_r)
MSE_r=sum_diffs_r/(len(y))

print()
print('regression model: IQ = ' + str(round(slope,2)) + ' * API ' + str(round(intercept, 2)))
print()
print('Mean Squared Error: '+ str(MSE))
print()
print('Point where IQ = 5.5: '+str(IQscore55))
print()
print('rotated regression model: IQ = ' + str(round(slope2,2)) + ' * API ' + str(round(intercept2, 2)))
print()
print('Mean Squared Error for rotated line: '+ str(MSE_r))
print()
```
